```python
Time_since_posted_in_hours=[]

for i in df_new['Time since posted']:
    Time_since_posted_in_hours.append(i.replace('hours', ''))
se = pd.Series(Time_since_posted_in_hours)
df_new['Times since posted (hours)'] = se.values
df_new['Times since posted (hours)'] = pd.to_numeric(df_new['Times since posted (hours)'])
df_new = df_new.drop('Time since posted', axis=1)
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
username=LabelEncoder()
```

```python
from sklearn.model_selection import train_test_split
X = df_new.drop(labels="Likes",axis=1)
y = df_new["Likes"]
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.33,random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error,r2_score
```

```python
import numpy as np
lr = LinearRegression()
lr.fit(X_train,y_train)
y_pred = lr.predict(X_test)

mse = mean_squared_error(y_test,y_pred)
rmse = np.sqrt(mse)
r2 = r2_score(y_test, y_pred)

print("Mean Squared Error (MSE):", mse)
print("Root Mean Squared Error (RMSE):", rmse)
print("R-squared (RÂ²):", r2)
```
